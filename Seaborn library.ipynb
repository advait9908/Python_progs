```python
import seaborn as sns
titanic = sns.load_dataset("titanic")
```

```python
titanic.shape
```

```python
titanic.keys()
```

```python
titanic.describe()
```

```python
# Vertical lines extending from boxes are called 'Whiskers'.
sns.boxplot(x='class',y='fare',data=titanic).set_title('Box plot for fare with respect to class')
```

```python
sns.violinplot(x='class',y='age',data=titanic).set_title('Violin plot for showing age with class and sex')
```

```python
sns.violinplot(x='class',y='age',hue='sex',data=titanic).set_title('Violin plot for showing age with class and sex')
```

```python
sns.pointplot(x='class',y='age',data=titanic).set_title('Point plot for showing age with respect to class')
```

```python
sns.lineplot(x='embarked',y='fare',data=titanic).set_title('Line plot for fare and embarked ')
```

```python
sns.countplot(y='alive',data=titanic).set_title('Count plot for alive')
```

```python
sns.barplot(x='embarked',y='fare',data=titanic).set_title('Bar plot for fare with respect to embarked ')
```

```python
sns.stripplot(x='class',y='fare',data=titanic).set_title('Strip plot for class and fare')
```

```python
sns.stripplot(x='class',y='fare', jitter=True, data=titanic).set_title('Strip plot for class and fare')
```

```python
sns.swarmplot(x='class',y='fare',data=titanic).set_title('Swarm plot for class and fare')
```

```python
sns.factorplot(x='sex',y='age',hue='alive',data=titanic)
```

```python
sns.catplot(x='sex',y='age',hue='alive',data=titanic)
```

```python
sns.catplot(x='class',y='fare',hue='alive',col='deck',col_wrap=3,kind='bar',data=titanic)
#col_wrap=3 means 3 columns in a row
```

```python
sns.FacetGrid(titanic,col='class')
```
